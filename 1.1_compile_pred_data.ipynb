# Training Data for Stacked Model

```python
# Compile and clean into one df

# Readability preds first
compiled_pred = readability_pred.copy()
compiled_pred.rename(columns={'helpful_pred':'readability_pred'}, inplace = True)
new_order = ['review_id', 'readability_pred']
compiled_pred = compiled_pred[new_order]

# Then review type
compiled_pred = compiled_pred.merge(review_type_pred[['review_id', 'helpful_predictions_few_shot']], on = 'review_id', how = 'outer')
compiled_pred.rename(columns={'helpful_predictions_few_shot':'review_type_pred'}, inplace = True)

# Then business nature
compiled_pred = compiled_pred.merge(bus_nat_pred[['review_id', 'helpful_predictions_final','helpful']], on = 'review_id', how = 'outer')
compiled_pred.rename(columns={'helpful_predictions_final':'business_type_pred'}, inplace = True)

# Then aspect
compiled_pred = compiled_pred.merge(aspect_pred[['review_id', 'helpful_pred']], on = 'review_id', how = 'outer')
compiled_pred.rename(columns={'helpful_predictions_final':'aspect_pred'}, inplace = True)
compiled_pred.rename(columns={'helpful_pred':'aspect_pred'}, inplace = True)
```

```python
compiled_pred = compiled_pred.dropna(subset=['helpful']) # drop rows with NaN in helpful col, ensure df only contains validation data
```

```python
# Convert review_type_pred to percentage
compiled_pred['review_type_pred'] = compiled_pred['review_type_pred'] * 100
```

```python

compiled_pred['aspect_pred'] = compiled_pred['aspect_pred'].fillna(0) # Fill with zeros, per Gia direction
```

```python
columns_to_update = ['readability_pred', 'review_type_pred', 'business_type_pred', 'aspect_pred']
compiled_pred[columns_to_update] = compiled_pred[columns_to_update].map(lambda x: max(x, 0))
```

```python
compiled_pred.head()
```

```python
compiled_pred.shape # should be 289061
```

Confirm no NaNs

```python
# Check which columns contain NaN values
columns_with_nans = compiled_pred.isna().any()

# Display columns with if NaN exists
print(columns_with_nans)
```

```python
# Compiled training data for stacked model
compiled_pred.to_parquet('/content/drive/MyDrive/Code + Data/stacked_model_predictions_train.parquet')
```

# Testing Data for Stacked Model

```python
# Compile and clean into one df for testing data

# Readability preds first
compiled_pred_test = readability_pred.copy()
compiled_pred_test.rename(columns={'helpful_pred':'readability_pred'}, inplace = True)
new_order = ['review_id', 'readability_pred']
compiled_pred_test = compiled_pred_test[new_order]

# Then review type
compiled_pred_test = compiled_pred_test.merge(review_type_pred[['review_id', 'helpful_predictions_few_shot']], on = 'review_id', how = 'outer')
compiled_pred_test.rename(columns={'helpful_predictions_few_shot':'review_type_pred'}, inplace = True)

# Then business nature
compiled_pred_test = compiled_pred_test.merge(bn_test_pred[['review_id', 'helpful_predictions_final','helpful']], on = 'review_id', how = 'outer')
compiled_pred_test.rename(columns={'helpful_predictions_final':'business_type_pred'}, inplace = True)

# Then aspect
compiled_pred_test = compiled_pred_test.merge(aspect_pred[['review_id', 'helpful_pred']], on = 'review_id', how = 'outer')
compiled_pred_test.rename(columns={'helpful_predictions_final':'aspect_pred'}, inplace = True)
compiled_pred_test.rename(columns={'helpful_pred':'aspect_pred'}, inplace = True)
```
